```python
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
df = pd.read_csv(r'stores_sales_forecasting.csv', encoding='ISO-8859-1')
df = df.sort_values(by='Order Date')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['Weekday'] = df['Order Date'].dt.weekday
df['Quarter'] = df['Order Date'].dt.quarter
```

```python
#Lag features
df['Sales Lag 1'] = df['Sales'].shift(1)
df['Sales Lag 12'] = df['Sales'].shift(12)
```

```python
#Rolling averages
df['Sales Rolling 3'] = df['Sales'].rolling(window=3).mean()
df['Sales Rolling 6'] = df['Sales'].rolling(window=6).mean()
```

```python
#Encode categorical variables
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['Region','State','Sub-Category']])
encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out())

df = pd.concat([df, encoded_df],axis=1)
df = df.dropna()
```

```python
#Feature selection
features = df[['Year','Month','Day','Weekday','Quarter','Sales Lag 1', 'Sales Lag 12', 'Sales Rolling 3','Sales Rolling 6'] + list(encoded_df.columns)] 

#Training and Validation data 
train_data = df.loc[df['Year']<2017]
val_data = df.loc[df['Year']==2017]

x_train = train_data[features.columns]
y_train = train_data['Sales']  #Fix later
x_val = val_data[features.columns]
y_val = val_data['Sales']
```

```python
#Choose and train suitable ML model
model = MLPRegressor(hidden_layer_sizes=(100,50),max_iter=500,random_state =42) 
#model = RandomForestRegressor(n_estimators=100, random_state=42)
#model = SVR(kernel='rbf',C=100,gamma=0.1,epsilon=.1)
#model = GradientBoostingRegressor(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
#Make predictions for 2017 and compare to values in validation set
y_pred = model.predict(x_val)
mse = mean_squared_error(y_val, y_pred)
print(f'Mean Squared Error: {mse}')
print("RMSE: " + str(round(math.sqrt(mse),2)))
```

```python
#Check error over entire months
val_data.loc[:,'Predicted Sales'] = y_pred
val_data.loc[:,'Actual Sales'] = y_val

monthly_predictions = val_data.groupby(['Year','Month']).agg({'Predicted Sales':'sum','Actual Sales':'sum'}).reset_index()

print(monthly_predictions.head(20))

monthly_mse = mean_squared_error(monthly_predictions['Actual Sales'], monthly_predictions['Predicted Sales'])

print(f'Monthly Mean Squared Error: {monthly_mse}')
print("RMSE: " + str(round(math.sqrt(monthly_mse),2)))

#Report overall error for 2017
print("The error for predicting 2017 total sales is: " + str(round(100* abs(val_data['Predicted Sales'].sum()-val_data['Actual Sales'].sum())/val_data['Actual Sales'].sum(),2)) + "%")
```

```python
#Plot results as scatter
plt.figure(figsize=(10, 6))
plt.scatter(val_data['Order Date'], y_val, label='Actual Sales')
plt.scatter(val_data['Order Date'], y_pred, label='Predicted Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Actual vs Predicted Sales in 2017')
plt.legend()
plt.show()
```
