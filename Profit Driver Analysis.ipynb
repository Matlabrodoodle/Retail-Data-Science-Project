```python
#The goal of this notebook is to determine if discounts drive annual profit trends in the store
#In particular, to explain why 2015 and 2017 had lower profits than 2014 and 2016 despite similar sales

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Import data
df = pd.read_csv(r'stores_sales_forecasting.csv', encoding='ISO-8859-1')
```

```python
#Check sum of profits by sale year
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year
yearly_profits = df.groupby("Year")["Profit"].sum()

plt.figure(figsize=(8, 5))
yearly_profits.plot(kind='bar', color='green')
plt.title('Total Profits by Sale Year')
plt.xlabel('Year')
plt.ylabel('Total Profits')
plt.xticks(rotation=0)
plt.show()
```

```python
#Check Profit margin also
df['Profit Margin'] = df['Profit']/df['Sales']*100

yearly_profit_margin = df.groupby('Year')['Profit Margin'].mean()

plt.figure(figsize=(8, 5))
plt.bar(yearly_profit_margin.index, yearly_profit_margin.values, color='orange')
plt.title('Mean Profit Margin by Sale Year')
plt.xlabel('Year')
plt.ylabel('Profit Margin')
plt.xticks(yearly_profit_margin.index.astype(int))
plt.show()
```

```python
#Determine transactions by month and year to look for seasonal trends and possible explanations for profit variability between years
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Year'] = df['Order Date'].dt.year

#Use format code %b to convert datetime to three-letter month abbreviation
df['Month'] = df['Order Date'].dt.strftime('%b')

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales = df.groupby(['Year', 'Month']).size().reset_index(name='Transactions')

# print(monthly_sales.head(10))

years = monthly_sales['Year'].unique()
monthly_sales['Month'] = pd.Categorical(monthly_sales['Month'], categories=months, ordered=True)

pivot_table = monthly_sales.pivot(index='Month', columns='Year', values='Transactions').reindex(months)

fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.2

r = np.arange(len(months))

for i, year in enumerate(pivot_table):
    ax.bar(r + i*bar_width,pivot_table[year],width=bar_width,edgecolor='grey',label=str(year))
    
ax.set_xlabel('Month')
ax.set_ylabel('Transaction Count')
ax.legend(title='Year')
ax.set_xticks(r + bar_width * (len(pivot_table.columns) - 1) / 2)
ax.set_xticklabels(months)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Hypothesis: frivolous discounts are responsible for lower profits in 2015 and 2017. Let's compare the mean discounts each year and see how they differ.

yearly_discounts = df.groupby("Year")["Discount"].mean()
plt.bar(yearly_discounts.index, yearly_discounts.values, color='orange')
plt.xticks(yearly_discounts.index.astype(int))

i = 2014

for value in yearly_discounts:
    print('Mean discount in ' + str(i) + ' is: ' + str(round(value,3)))
    i += 1

plt.show()
```

```python
#The discounts do not appear to be the reason for the fluctuations in profits.
#Hypothesis: the quantities of items sold per transaction cause the observed annual variation in profits

yearly_mean_quantities = df.groupby("Year")["Quantity"].mean()
plt.bar(yearly_mean_quantities.index, yearly_mean_quantities.values,color='purple')
plt.xticks(yearly_mean_quantities.index.astype(int))

i = 2014

for value in yearly_mean_quantities:
    print('Mean quantity sold per transaction in ' + str(i) + ' is: ' + str(round(value,3)))
    i += 1

plt.show()
```

```python
#Again no dice, there is no explanation above for why profits fluxuate by year as much as they do.
#Hypothesis: the mean sales values are lower in years with lower profits

yearly_mean_sales = df.groupby("Year")["Sales"].mean()

i = 2014

for value in yearly_mean_sales:
    print('Mean sales value in ' + str(i) + ' is: ' + str(round(value,3)))
    i += 1

plt.bar(yearly_mean_sales.index,yearly_mean_sales.values,color='gray')
plt.xticks(yearly_mean_sales.index.astype(int))
plt.show()
```

```python
#Mean sales values do not appear to explain profit trends either
#Hypothesis: years with higher profits simply have more total transactions

Transactions_Yearly_Count = df["Year"].value_counts().sort_index()

i = 2014

for value in Transactions_Yearly_Count:
    print('Overall transaction count in ' + str(i) + ' is: ' + str(round(value,3)) + ' transactions.')
    i += 1

plt.bar(Transactions_Yearly_Count.index, Transactions_Yearly_Count.values,color='purple')
plt.xticks(yearly_discounts.index.astype(int))
plt.show()
```

```python
#Still no clear explanation for profit trends
#Introduce new series for unit price (not present in original data)

df["Unit Price"] = df["Sales"]/(df["Quantity"]*(1-df["Discount"]))
df["Unit Price"].describe()
```

```python
#Hypothesis: goods sold in lower-profit years have lower individual prices (although that should show up in Sales too)
Price_by_year = df.groupby("Year")["Unit Price"].mean()
plt.bar(Price_by_year.index, Price_by_year.values,color='blue')  
plt.xticks(yearly_discounts.index.astype(int))
plt.show()
```
